## Two-body Relative Dynamics

$$
\bm{\ddot r} = -\frac{\mu}{r^3}\bm{r}
$$

- System of 3 scalar second order differential equations
- It can be reduced to a system of 6 $\textcolor{red}{first}$ order equations

$$ \bm{\dot r} = \bm{v} $$
$$ \bm{\dot v} = -\frac{\mu}{r^3}\bm{r} $$

- To solve this system we need 6 initial conditions with $\bm{r_0}$ and $\bm{v_0}$ as known intial values

$$ \bm{r}(t_0) = \bm{r_0}$$
$$ \bm{v}(t_0) = \bm{v_0}$$

- The system admits at maximum 6 independent integrals of motion

> An Integral of Motion is a function $f(\bm{r},\bm{v},t)$ that is constant for all times, *t*

$$ f(\bm{r},\bm{v},t) = const
$$

- The value of the constant is determined by the intial conditions
- An integral of motion reduces the degree of freedom of our problem


## Angular Momentum 

- The relative angular momentum of $m_2$ with respect $m_1$ is 
$${\bf H}_{2/1} = \bm{r}\times(m_2\bm{v}) = \bm{r}\times(m_2\bm{\dot r})  $$

- per unit mass gives the specific angular momentum
$$ \bm{h}=\frac{{\bf H}_{2/1}}{m_2} = \bm{r}\times\bm{\dot r}$$

- the time derivative of $\bm{h}$ is 
$$ \frac{d\bm{h}}{dt} =\frac{d}{dt}(\bm{r}\times\bm{\dot r}) = \bm{\dot r}\times\bm{\dot r} + \bm{r}\times\bm{\ddot r} =  \bm{r}\times\bm{\ddot r}
$$

- From the dynamics $\bm{\ddot r}=-\frac{\mu}{r^3}\bm{r}$, thus 
$$\frac{d\bm{h}}{dt}=\bm{r}\times\bm{\ddot r}=\bm{r}\times-\frac{\mu}{r^3}\bm{r}=0
$$

$$
\frac{d\bm{h}}{dt}=0
$$


$$
\bm{h} = const
$$

This result has the following implications:

- $\bm{h}(\bm{r},\bm{\dot r},t) = const$
- $\bm{h}$ is an integral of motion
- All 3 scalar quantities are constants: $h_x(\bm{r},\bm{\dot r},t) = const$, $h_y(\bm{r},\bm{\dot r},t) = const$, $h_z(\bm{r},\bm{\dot r},t) = const$

## Consequences of $\bm{h} =const$

$$\bm{h} = \bm{r}\times\bm{v}$$

hence 

$$ \bm{h}\perp\bm{r}\times\bm{v} $$

**Direction** 



- $\bm{r}$ and $\bm{v}$ must at all times lie in a plane perpendicular to $\bm{h}$ 
- The motion is planar

**Magnitude**



$$ \bm{h} = \bm{r}\times\bm{v} = \bm{r}\times(v_r\bm{e_r} + v_\theta\bm{e_\theta})$$

$$=\bm{r}\times(v_r\bm{e_r}) + \bm{r}\times(v_\theta\bm{e_\theta})$$

$$=rv_\theta\bm{\hat{h}}$$

> $rv_\theta =const$

$$dA = \frac{1}{2}rvdt\sin(\phi)$$
$$\frac{dA}{dt} = \frac{1}{2}rv\sin(\phi) = \frac{1}{2}rv_\theta$$

> $\frac{dA}{dt}=\frac{1}{2}h = const$

> *Kepler's Second Law: Areal velocity is constant*

## Polar Coordinates

If the motion is planar we can use polar coordinates to fully describe it. 



- In polar coordinates $\bm{r} = r\bm{e_r}$

$$\frac{d\bm{r}}{dt} = \bm{v} = \bm{\dot r}\bm{e_r} + \bm{r}\bm{e_\theta}\dot \theta $$

$$\bm{h}=\bm{r}\times\bm{v}=r\bm{e_r}\times(\bm{\dot r}\bm{e_r}+\bm{r}\bm{e_\theta}\dot \theta)
$$

$$=r^2\dot \theta\bm{k}=rv_\theta\bm{k}=const
$$

>$$\frac{dA}{dt}=\frac{1}{2}h=\frac{1}{2}r^2\dot \theta$$

$$\frac{d^2\bm{r}}{dt}=\bm{a}=(\bm{\ddot r}-r\dot \theta^2)\bm{e_r}+(r\ddot \theta+2\dot r\dot \theta)\bm{e_\theta}$$

$$\frac{-\mu}{r^2}\bm{e_r}
$$

1. $$\bm{\ddot r}- r\dot \theta^2=\frac{-\mu}{r^2}$$
2. $$r\ddot \theta + 2\bm{\dot r}\dot \theta = 0\Leftrightarrow  \frac{d}{dt}(r^2\dot \theta)=2r\bm{\dot r}\dot \theta+r^2\ddot \theta=r(r\ddot \theta+2\bm{\dot r}\dot \theta) =0$$



Eqiation 2 is equivalent to $\bm{h}=const$ which is a scalar

## Conservation of Specific Orbital Energy: $E$

Take the two-body dynamics and the dot product with $\bm{\dot r}$
$$
\bm{\dot r}\cdot\bm{\ddot r}=\frac{-\mu}{r^2}\bm{\hat{r}}\cdot\bm{\dot r}
$$

1. Note that 
$$ \frac{d}{dt}(\frac{1}{2}v^2)=\frac{1}{2}\frac{d}{dt}(\bm{\dot r\cdot\dot r})
$$

$$\frac{1}{2}\bm{\ddot r}\cdot\bm{\dot r}+\frac{1}{2}\bm{\dot r}\cdot\bm{\ddot r}=\bm{\dot r}\cdot\bm{\ddot r}$$

thus 

- $$\bm{\dot r}\cdot\bm{\ddot r}=\frac{d}{dt}(\frac{1}{2}v^2)$$

>Notice that this is the derivative of the kinetic energy

- In polar coordinates $v^2=\bm{v_r}^2+\bm{v_\theta}^2=\bm{\dot r}^2+r^2\dot \theta^2$

2. Note that 
$$\frac{d}{dt}(\frac{\mu}{r})=\frac{\partial}{\partial r}\frac{\mu}{r}\frac{dr}{dt}=-\frac{\mu}{r^2}\bm{\dot r}$$

thus 

$$-\frac{\mu}{r^2}\bm{\hat{r}}\cdot\bm{\dot r}=\frac{\mu}{r^2}\bm{\dot r}=\frac{d}{dt}(\frac{\mu}{r})$$

>Notice that this is the derivative of the potential of the gravotational force

1 + 2 

$$\frac{d}{dt}(\frac{1}{2}v^2-\frac{\mu}{r})=0
$$

$$\frac{1}{2}v^2-\frac{\mu}{r}=const$$

>Notice that this is the conservation of specific orbital energy

## Admissible Orbital Radius

$$
E=\frac{1}{2}v^2-\frac{\mu}{r}=\frac{1}{2}\bm{\dot r}^2+\frac{1}{2}r^2\dot \theta-\frac{\mu}{r}=\frac{1}{2}\bm{\dot r}^2+\frac{1}{2}\frac{h^2}{r^2}-\frac{\mu}{r}
$$

- $\frac{1}{2}\frac{h^2}{r^2}-\frac{\mu}{r}$ is the effective potential $\phi(r)$
- $\frac{1}{2}\frac{h^2}{r^2}$ is the potential of the contributed force
- $\frac{\mu}{r}$ is the potential gravitational force



The motion is only possible for those values of $r$ such that $E\geq\phi_{eff}$



- For $E<0$ the motion occurs between $r_{min}-r_{max}$
- For $E\geq0$ the motion is unbounded
- For $E=min\phi(r), r_{min}=r_{max}$ hence we have a circular orbit
