## Hyperbolic Orbit


- When $e>1$ the equation of the orbit describes a hyperbola

```{math} 
:label:L8_1
r=\frac{h^2/\mu}{1+e\cos\theta}
```

- The orbit radius must be $>0$ thus $1+e\cos\theta\ge0\Rightarrow\cos\theta\ge-\frac{1}{e}$ for $-\theta_{Max}\le\theta\le\theta_{Max}$

- When $\theta=\theta_{Max}\Rightarrow r\rightarrow\infty$, we call $\theta_{Max}=\theta_{\infty}$ 

```{math} 
:label:L8_2 
\theta_{\infty}=\arccos\frac{1}{e}
```

- When $r=r_{\infty}$, $v=v_{\infty}$ and 

```{math} 
:label:L8_3 
\frac{1}{2}v_{\infty}^2 - \frac{\mu}{r_{\infty}} =E
``` 

but 

```{math} 
:label:L8_4
\frac{\mu}{r_{\infty}} = 0
``` 

hence 

```{math} 
:label:L8_5 
 E= \frac{1}{2}v_{\infty}^2\ge0
```

- The hyperbola is the locus of points for which the difference of the distance from the two foci is constant $=2a$
- If we work with negative $a$ we can use the same formulae of the ellipse

```{math} 
:label:L8_6 
\rightarrow\theta=90\degree\Rightarrow r=p=\frac{h^2}{\mu}=a(1-e^2)
```

```{math} 
:label:L8_7 
\rightarrow\theta=0\degree\Rightarrow r_p=\frac{\frac{h^2}{\mu}}{1+e}=\frac{p}{1+e}=a(1-e)
```

```{math} 
:label:L8_8 
\rightarrow E=\frac{1}{2}v_{\infty}^2=-\frac{\mu}{2a}
```

## Interplanetary Missions: Patched Conics Model


- Preliminary mission desing uses patched conics approximation 
1. The whole transfer is split into legs described with two-body dynamics 
2. The planets' sphere of influences are infinitely large in planetary reference frame but points in heliocentric reference frame 
3. Planetary legs are instantaneous hyperbolic arcs
4. Different legs are patched together either by gravity assist maneuvers or impulse $\Delta v$

![Figure 2](images/9.png)

- $1$ and $2$ are heliocentric arcs (two-body problem under the gravitational pull of the sun)
- If we zoom in at Earth's departure, Mars encounter and Jupiter arrival we see hyperbolic arcs in which the spacecraft feels the gravitational pull only of the Earth, Mars and Jupiter respectively
- At Mars we perform a **Gravity Assist Maneuver**:

> A close passage with a celestial body (in this case it's Mars) is exploited to change the heliocentric velocity of the spacecraft from
$\bf{V^-}$ to $\bf{V^+}$: a $\Delta \bf{v} = \bf{V^+}-\bf{V^-}$ for free!

## Gravity Assist



![Figure 3](images/10.png)

![Figure 4](images/11.png)

- $\bf{V_M}$ is the heliocentric velocity of Mars
- $\bf{v_{\infty}^-} = \bf{V^-} -  \bf{V_M}$ is the incoming velocity vector $\underline{relative}$ to Mars
- $\bf{v_{\infty}^+} = \bf{V^+} -  \bf{V_M}$ is the outgoing velocity vector $\underline{relative}$ to Mars
- As the planetary sphere of influence is infinitely large it follows, hence a hyperbolic arc!

```{math} 
:label:L8_9 
\frac{1}{2}(v_{\infty}^-)^2 - \frac{\mu}{r_{\infty}} = \frac{1}{2}(v_{\infty}^-)^2>0 \Rightarrow E>0
```

> Note $\frac{\mu}{r_{\infty}} = 0$
- In the two-body problem: 

```{math} 
:label:L8_10 
E=const \Rightarrow \frac{1}{2}(v_{\infty}^-)^2 - \frac{\mu}{r_{\infty}} = \frac{1}{2}(v_{\infty}^+)^2 - \frac{\mu}{r_{\infty}}
```

```{math} 
:label:L8_11
\frac{\mu}{r_{\infty}} = 0
```

```{math} 
:label:L8_12
\Rightarrow v_\infty^+ = v_\infty^-
```

- The absolute velocity $\bf{V^-}$ and $\bf{V^+}$ are different because the gravity assist steers the velocity $\bf{v_{\infty}^-}$
- The turn angle $\delta$ determines the velocity change $\Delta\bf{v}=\bf{V^+}-\bf{V^-}$
> We obtain $\Delta\bf{v}$ without propellent consumption


![Figure 5](images/12.png)

![Figure 6](images/13.png)

- the turn angle $\delta$ determines the velocity change 
> NOTE: 

```{math} 
:label:L8_13
\theta_{\infty}=90\degree + \delta
```

and 

```{math} 
:label:L8_14
1+e\cos\theta_{\infty}=0 \Rightarrow 1+e\cos(90+\delta)=0 \Rightarrow 1-e\sin\delta = 0
```

> 
```{math} 
:label:L8_15
\sin\delta=\frac{1}{e}
```

>1. We get a parabola when $e\rightarrow 1$ , $\delta\Rightarrow 90\degree \Rightarrow 2\delta \rightarrow 180\degree$ maximum turn! 
>2. $e\rightarrow \infty \Rightarrow \delta \rightarrow 0 \Rightarrow$ minimum turn!

- With $v_\infty$ and $e$ the gravity assist is fully determined

```{math} 
:label:L8_16
\rightarrow \frac{1}{2}v_{\infty}^2 - \frac{\mu}{r_{\infty}} = -\frac{\mu}{2a}
```

```{math} 
:label:L8_17
\frac{\mu}{r_{\infty}} = 0
```

```{math} 
:label:L8_18
\Rightarrow a=-\frac{\mu}{v_{\infty}^2}
```

```{math} 
:label:L8_19
r_p=a(1-e)=-\frac{\mu}{v_{\infty}^2}(1-e)\Rightarrow e=1+\frac{r_{p}v_{\infty}^2}{\mu}
```

>$r_p$ is the main design parameter

### To Maximise $\delta$
- Low $r_p$ (limited by planetary radius)
- Low $v_{\infty}$
- High $\mu$ (massive planet!)

#### Velocity at perigee $v_p$

```{math} 
:label:L8_20
\frac{1}{2}v_{p}^2 - \frac{\mu}{r_{p}} = \frac{1}{2}v_{\infty}^2
```

```{math} 
:label:L8_21  
r_p=\frac{\mu}{v_{\infty}^2}(e-1)
```

```{math} 
:label:L8_22
\Rightarrow v_p^2=v_{\infty}^2+\frac{2\mu v_{\infty}^2}{\mu (e-1)} \Rightarrow v_p=v_{\infty}\sqrt{\frac{e-1+2}{e-1}} 
```

```{math} 
:label:L8_23
\Rightarrow v_p=v_{\infty}\sqrt{\frac{e+1}{e-1}} 
```

#### Sometimes it is more practical to work with the aiming distance $\Delta$ as design parameter


![Figure 7](images/14.png)

```{math} 
:label:L8_24
h=v_{\infty}\Delta = r_p v_p
```

```{math} 
:label:L8_25
\rightarrow r_p=\frac{\mu}{v_{\infty}^2}(e-1)
```

```{math} 
:label:L8_26
\rightarrow v_p = v_{\infty}\sqrt{\frac{e+1}{e-1}}
```

```{math} 
:label:L8_27
h = \frac{\mu}{v_{\infty}^2} (e-1) \cdot v_{\infty}\sqrt{\frac{e+1}{e-1}} = \frac{\mu}{v_{\infty}} \sqrt{e^2-1}
```

```{math} 
:label:L8_28
\Rightarrow \Delta = \frac{\mu}{v_{\infty}^2}\sqrt{e^2-1}
``` 

```{math} 
:label:L8_29
e^2=1+\frac{\Delta^2 v_{\infty}^4}{\mu^2} 
```


### To Maximise $\delta$
- Low $\Delta$ (limited by planetary radius)
- Low $v_{\infty}$
- High $\mu$ (massive planet!)
